```python
def detectFaceSSD(image, isGroup = 0):
    w = image.shape[1]
    h = image.shape[0]
    
    inWidth = 100
    inHeight = 100
    
    if(isGroup):
        inWidth = 640
        inHeight = 640
    
    blob = cv2.dnn.blobFromImage(image, 1.0, (inWidth, inHeight), (104.0, 177.0, 123.0), False, False)
    net.setInput(blob)
    detections = net.forward()
    count = 0
    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        if(confidence > 0.5):
            count = count + 1
            xLeftBottom = int(detections[0, 0, i, 3] * w)
            yLeftBottom = int(detections[0, 0, i, 4] * h)
            xRightTop = int(detections[0, 0, i, 5] * w)
            yRightTop = int(detections[0, 0, i, 6] * h)

            cv2.rectangle(image, (xLeftBottom, yLeftBottom), (xRightTop, yRightTop), (255, 0, 0), 3)
            label = "Face: %.4f" % confidence
            labelSize, baseLine = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 1)

            cv2.rectangle(image, (xLeftBottom, yLeftBottom - labelSize[1]),
                          (xLeftBottom + labelSize[0], yLeftBottom + baseLine),
                          (255, 0, 0), cv2.FILLED)

            cv2.putText(image, label, (xLeftBottom, yLeftBottom),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0))
    print(count)
    return image
```
